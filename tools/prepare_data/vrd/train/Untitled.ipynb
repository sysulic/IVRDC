```python
import cPickle as cp
import numpy as np

f = open('train_gt_file.pkl', 'rb')
dic = cp.load(f)
all_train_gts = dic['gt_label']
all_train_gt_bboxes = dic['gt_box']
f.close()

print all_train_gt_bboxes[0]
```

```python
aaa = []
for im in all_train_gts:
#     print(im)
    for tri in im:
#         print(tri)
        if tri not in aaa:
            aaa.append(tri)
print(aaa)
```

```python
with open('aaa.txt','w')as fid:
    for a in aaa:
        fid.write("%s\t%s\t%s\n"%(str(a[0]),str(a[1]),str(a[2])))
```
